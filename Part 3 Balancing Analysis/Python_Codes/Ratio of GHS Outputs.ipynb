```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from a CSV file
# (Update this path if running on another computer or after uploading to GitHub)
df = pd.read_csv(r"C:\Users\user\Desktop\u\Y3S1\ZCT 398\02. data\Full data.csv")

# Extract the "Pictogram(s)" column and remove missing values (NaN)
# Some cells may contain multiple pictograms separated by commas
pictograms = df["Pictogram(s)"].dropna().str.split(",")

# Convert list-like entries into separate rows
# Example: "flame, skull" becomes two separate rows
pictograms = pictograms.explode().str.strip()

# Count how many times each GHS pictogram appears
pictogram_counts = pictograms.value_counts().reset_index()

# Rename columns for clarity
pictogram_counts.columns = ["GHS", "Count"]

# Display the result in the console
print(pictogram_counts)

# Save the results to a new CSV file for future use or reporting
pictogram_counts.to_csv("pictogram_counts.csv", index=False)

# ------------------- Visualization Section -------------------

# Create a figure with a specific size
plt.figure(figsize=(10, 6))

# Create a bar chart showing frequency of each GHS pictogram
plt.bar(pictogram_counts["GHS"], pictogram_counts["Count"])

# Label x-axis and y-axis
plt.xlabel("GHS")
plt.ylabel("Count")

# Add a title to the plot
plt.title("Count of Each GHS")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Adjust layout to prevent overlapping elements
plt.tight_layout()

# Save the plot as a high-quality JPG image
plt.savefig("pictogram_counts.jpg", dpi=300, format="jpg")

# Display the plot
plt.show()
```
